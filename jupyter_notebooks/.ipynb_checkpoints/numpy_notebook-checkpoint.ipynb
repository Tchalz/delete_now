```python
sales = np.genfromtxt("sales.csv", delimiter = ",", dtype = "|U3") #read-in text data as string
sales
    
    
```

```python
sales = np.genfromtxt("sales.csv", delimiter = ",", dtype = "int") #read-in text data as string
sales
    
    
```

```python
sales_numbers = sales[1:, :]
sales_numbers
```

```python
year1_sales = sales_numbers = sales[1:53, :]
sales_numbers
```

```python
monday_sales = year1_sales[:,1]
monday_sales
```

```python
tuesday_sales = year1_sales[:, 2]
tuesday_sales
```

```python
monday_sales_amount_usd = monday_sales * 10
monday_sales_amount_usd
```

```python
month1_year1_sales = year1_sales[:4,1:]
days = ["monday","tuesday","wednesay","thursday","friday","saturday","sunday"]
month1_year1_sales
mean_sales = []
for i in range(month1_year1_sales.shape[1]):
    day_data = month1_year1_sales[:,i]
    mean_sales.append(day_data.mean())
    print(days[i], day_data, f"mean sales ===>   {day_data.mean()}")
```

```python
plt.bar(days, mean_sales)
```
