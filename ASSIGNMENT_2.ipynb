##SECTION 2: DATA CLEANING

###QUESTION 7(i): Handling missing values

#####TOTAL_BEDROOMS MISSING VALUES

```python
total_bedrooms_median = df['total_bedrooms'].median()
total_bedrooms_median
```

```python
df['total_bedrooms'] = df['total_bedrooms'].fillna(total_bedrooms_median)
```

(1) For the total_bedrooms missing values, I chose to fill it up with median.

(2) Because of Outliers that exist in the column. Median is usually not affected by extreme values

(3) Sanity check to confirm there are no missing values in total_bedrooms

```python
print(df['total_bedrooms'].isnull().sum())
```

#####POPULATION MISSING VALUES

```python
population_median = df['population'].median()
population_median
```

```python
df['population'] = df['population'].fillna(population_median)
```

(1) For the population missing values, I chose to fill it up with median.

(2) Because of Outliers that exist in the column. Median is usually not affected by extreme values



(3) Sanity check to confirm there are no missing values in population column

```python
print(df['population'].isnull().sum())
```

#####MEDIAN_INCOME MISSING VALUES

```python
median_income_mean = df['median_income'].mean()
print(median_income_mean)
```

```python
df['median_income'] = df['median_income'].fillna(median_income_mean)
```

(1) For the median_income missing values, I chose to fill it up with mean.

(2) Because there are no Outliers that exist in the column, all the datasets are used here and the precise average is calculated.

(3) Sanity check to confirm there are no missing values in median_income column



```python
print(df['median_income'].isnull().sum())
```

##SECTION 3: FEATURE SCALING

#####QUESTION 9 ANSWERS

(i) Feature scaling is a technique used in machine learning to bring all numerical features to a similar range of values or scale.

(ii) The problem that will arise without feature scaling is that When one variable has very large numbers (like income) and another has much smaller values (like age), the model may wrongly focus more on the large valued feature, even though it is not actually more important than the other.

##QUESTION 10 ANSWERS

#####question 10, 11(1)

```python
col = 'housing_median_age'
col_min = df[col].min()
col_max = df[col].max()
df[col + '_scaled'] = (df[col] - col_min) / (col_max - col_min)
print(df[col + '_scaled'].min())
print(df[col + '_scaled'].max())
```

#####qusetion 10, 11(2)

```python
col = 'total_rooms'
col_min = df[col].min()
col_max = df[col].max()
df[col + '_scaled'] = (df[col] - col_min) / (col_max - col_min)
print(df[col + '_scaled'].min())
print(df[col + '_scaled'].max())
```
