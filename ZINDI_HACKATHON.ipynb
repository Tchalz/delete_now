```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
train=pd.read_csv(r'/Train.csv')
train
```

```python
train.fillna(method="bfill", inplace=True)
train
```

```python
train['Year']=train['Year'].str.replace(',', '').astype(int)
train['Distance']=train['Distance'].str.replace(',', '').astype(float)
```

```python
train
```

```python
test=pd.read_csv(r'/Test.csv')
test
```

```python
test.fillna(method="ffill", inplace=True)
test
```

```python
test['Year']=test['Year'].str.replace(',', '').astype(int)
test
```

```python
y=train['Amount (Million Naira)']
X=train.drop(['Amount (Million Naira)', 'Model'], axis=1)
X_test=test.drop(['VehicleID'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cat_cols=['VehicleID', 'Location', 'Maker', 'Colour', 'Type']
for cols in cat_cols:
    X[cols]=le.fit_transform(X[cols])
print(X)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cat_cols=['Location', 'Maker', 'Model', 'Colour', 'Type']
for cols in cat_cols:
    X_test[cols]=le.fit_transform(X_test[cols])
print(X_test)
```

```python
from sklearn.svm import SVR
sv_reg=SVR()
sv_reg.fit(X, y)
prediction=sv_reg.predict(X_test)
prediction
```

```python
df = pd.DataFrame({'VehicleID':test.VehicleID, 'Amount (Million Naira)':prediction})
```

```python
df.to_csv('submission_sv2.csv')
```
